```python
#with referrence to https://www.researchgate.net/publication/321057905_Emoji_as_Emotion_Tags_for_Tweets
#https://github.com/leyaoliatan/Sentiments-in-Tweets-with-Emojis/blob/main/1_tweet_collect_en.ipynb
joy = ['\U0001F600', '\U0001F602', '\U0001F603', '\U0001F604',
          '\U0001F606', '\U0001F607', '\U0001F609', '\U0001F60A',
          '\U0001F60B', '\U0001F60C', '\U0001F60D', '\U0001F60E',
          '\U0001F60F', '\U0001F31E', '\U0000263A', '\U0001F618',
          '\U0001F61C', '\U0001F61D', '\U0001F61B', '\U0001F63A',
          '\U0001F638', '\U0001F639', '\U0001F63B', '\U0001F63C',
          '\U00002764', '\U0001F496', '\U0001F495', '\U0001F601',
          '\U00002665']#joy

anger = ['\U0001F62C', '\U0001F620', '\U0001F610',
          '\U0001F611', '\U0001F620', '\U0001F621', '\U0001F616',
          '\U0001F624', '\U0001F63E']#anger
disgust= ['\U0001F4A9']#disgust
fear = ['\U0001F605', '\U0001F626', '\U0001F627', '\U0001F631',
          '\U0001F628', '\U0001F630', '\U0001F640']#fear
sad = ['\U0001F614', '\U0001F615', '\U00002639', '\U0001F62B',
          '\U0001F629', '\U0001F622', '\U0001F625', '\U0001F62A',
          '\U0001F613', '\U0001F62D', '\U0001F63F', '\U0001F494']#sad
surp = ['\U0001F633', '\U0001F62F', '\U0001F635', '\U0001F632']#surprise
happy=joy+surp
sad=disgust+fear+sad+anger
```

```python
def countsad (a):
  count=0
  for i in a :
    if i in sad:
      count+=1
  return count
def counthappy ( a):
  count=0
  for i in a:
    if i in happy:
      count+=1
  return count 



```

```python
df['sad']=df['text'].apply(countsad)
df['happy']=df['text'].apply(counthappy)
```

```python
df.tail(10)
```

```python
def final(sad, happy, pol):#pol:polarity
    if sad >0 and happy==0:
      return -1
    elif happy>0 and sad==0:
      return 1
    elif pol>0:
      return 1
    elif pol<0 :
      return -1
    else:
       return 0

```

```python
df['sentiment-overall']=df.apply(lambda row : final(row['sad'],
                     row['happy'], row['calculated']), axis = 1)
```

#preprocessing of tweets


Removing # and @ symbols and hastags  also removing RT(retweet) and urls attached



```python
import re
def cleantweets(text):
  text=re.sub(r'@[A-Za-z0-9]+',"",text) 
  text=re.sub(r'#',"",text)  ##remove
  text=re.sub(r'RT[\s]+',"",text) #retweets remove
  text=re.sub(r'https?:\/\/\S+',"",text) #urls remove
  return text

df['text']=df['text'].apply(cleantweets)  
df
```

##removing punctuations

```python
import string
string.punctuation # contains all possible punctuations

def remove_punct(text):
    text  = "".join([char for char in text if char not in string.punctuation])
    text = re.sub('[0-9]+', '', text)
    return text

df['text'] = df['text'].apply(lambda x: remove_punct(x))
df.head(10)
```

```python
import string
string.punctuation # contains all possible punctuations

def remove_punct(text):
    text  = "".join([char for char in text if char not in string.punctuation])
    text = re.sub('[0-9]+', '', text)
    return text

df['text'] = df['text'].apply(lambda x: remove_punct(x))
df.head(10)
```

### converting emojis to text (only happy and sad list)

```python
dhappy={}
dsad={}
for uni in happy:
  dhappy[uni]='happy'
for uni in sad:
  dsad[uni]='sad'

print(dhappy,dsad)    


def repuni(x):
  for key, value in dhappy.items():
    # Replace key character with value character in string
    x = x.replace(key, value)
  for key, value in dsad.items():
    # Replace key character with value character in string
    x = x.replace(key, value)
  return x  
```

```python
df['text']=df['text'].apply(repuni)    
```

##Tokenisation using nltk

```python
import nltk
nltk.download('punkt')
from nltk.tokenize import word_tokenize
df['tokenized_sents'] = df.apply(lambda row: word_tokenize(row['text']), axis=1)
```

## Lemmitisation using nltk


```python
nltk.download('wordnet')
nltk.download('omw-1.4')
wn = nltk.WordNetLemmatizer()

def lemmatizer(text):
    text = [wn.lemmatize(word) for word in text]
    return text

df['lemmatized'] = df['tokenized_sents'].apply(lambda x: lemmatizer(x))
df.head()
```

## generation of unigrams and bigrams to use them as a feature 
###with reference to https://www.tutorialspoint.com/python_text_processing/python_bigrams.htm

```python
##joining the lemmitized text 

df['lemmatized']=df['lemmatized'].apply(lambda x: " ".join(x))
df
```

### Creation of unigram

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfuni = TfidfVectorizer(analyzer='word',ngram_range=(1, 1), stop_words='english')
tfidf_matrix = tfuni.fit_transform(df['lemmatized'])
```

```python

tfidf_matrix.shape
dfunigram=pd.DataFrame.sparse.from_spmatrix(tfidf_matrix,columns=tfuni.get_feature_names_out())
dfunigram['label']=df['sentiment-overall']
```

### Creation of bigram and storing them in data frame

```python

tfbi = TfidfVectorizer(analyzer='word',ngram_range=(2, 2), stop_words='english')
tfidf_matrix1 = tfbi.fit_transform(df['lemmatized'])
```

```python
tfidf_matrix1.shape
dfbigram=pd.DataFrame.sparse.from_spmatrix(tfidf_matrix1,columns=tfbi.get_feature_names_out())
dfbigram['label']=df['sentiment-overall']
```

##Applying Naive Bias and SVM over unigrams and bigram
### first applying naive bias over unigrams 

```python
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score,f1_score,confusion_matrix

X_train1, X_test1, y_train1, y_test1 = train_test_split(dfunigram.iloc[:,:-1], dfunigram['label'], test_size = 0.20, random_state = 32)
nbuni = MultinomialNB()
nbuni.fit(X_train1,y_train1)
y_pred1  = nbuni.predict(X_test1)
print("Score",accuracy_score(y_pred1,y_test1))
print("F1-score",f1_score(y_pred1,y_test1,average='weighted'))
```

### Applying Naive Bias over Bigrams 

```python
X_train2, X_test2, y_train2, y_test2 = train_test_split(dfbigram.iloc[:,:-1], dfbigram['label'], test_size = 0.20, random_state = 32)
nbbig = MultinomialNB()
nbbig.fit(X_train2,y_train2)
y_pred2  = nbbig.predict(X_test2)
print("Score",accuracy_score(y_pred2,y_test2))
print("F1-score",f1_score(y_pred2,y_test2,average='weighted'))
```

###Applying Svm over unigrAM

```python
from sklearn.svm import LinearSVC 
svuni=LinearSVC()
svuni=svuni.fit(X_train1,y_train1)
y_pred3 =svuni.predict(X_test1)
print("Score",accuracy_score(y_pred3,y_test1))
print("F1-score",f1_score(y_pred3,y_test1,average='weighted'))
```

### Applying Svm over bigram

```python
svbi=LinearSVC()
svbi.fit(X_train2,y_train2)
y_pred4 =svbi.predict(X_test2)
print("Score",accuracy_score(y_pred4,y_test2))
print("F1-score",f1_score(y_pred4,y_test2,average='weighted'))
```
